```python
# Standardize the data with StandardScaler().
X_scaled = StandardScaler().fit_transform(wine_df_2)
print(X_scaled[0:5])
```

```python
# Using PCA to reduce dimension to three principal components.

# Initialize PCA model
pca = PCA(n_components=3)

# fit and transform the scaled data
wine_pca = pca.fit_transform(X_scaled)
```

```python
# Create a DataFrame with the three principal components.

pca_df = pd.DataFrame(
    data=wine_pca, columns=["PC 1", "PC 2", "PC 3"], index = wine_df_2.index
)
pca_df.head(10)
```

```python
# Create an elbow curve to find the best value for K.
# Finding the best value for k
inertia = []
k = list(range(1, 13))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pca_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

# Plot the curve
# df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)

# Plotly express plot
fig = px.line(df_elbow, x="k", y="inertia", title='Elbow Curve')
fig.show()
```

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pca_df)

# Predict clusters
predictions = model.predict(pca_df)
```

```python
#Create a new DataFrame including predicted clusters and features.
# Concatentate the DataFrames on the same columns.
clustered_df = pd.concat([wine_df_2, pca_df], axis = 1)
#clustered_df.head()

#  Add a new column that holds the predictions.
clustered_df["Clusters"] = model.labels_

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.sample(10)
```

```python
# Creating a 3D-Scatter with the PCA data and the clusters

# Plot 3D
fig = px.scatter_3d(
    clustered_df,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    color="Clusters",
       
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
# Initialize the K-Means model.
model2 = KMeans(n_clusters=3, random_state=0)

# Fit the model
model2.fit(pca_df)

# Predict clusters
predictions2 = model2.predict(pca_df)
```

```python
#Create a new DataFrame including predicted clusters and features.
# Concatentate the DataFrames on the same columns.
clustered_df2 = pd.concat([wine_df_2, pca_df], axis = 1)
#clustered_df.head()

#  Add a new column that holds the predictions.
clustered_df2["Clusters"] = model2.labels_

# Print the shape of the clustered_df
print(clustered_df2.shape)
clustered_df2.sample(10)
```

```python
# Creating a 3D-Scatter with the PCA data and the clusters

# Plot 3D
fig2 = px.scatter_3d(
    clustered_df2,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    color="Clusters",
       
)
fig2.update_layout(legend=dict(x=0, y=1))
fig2.show()
```
