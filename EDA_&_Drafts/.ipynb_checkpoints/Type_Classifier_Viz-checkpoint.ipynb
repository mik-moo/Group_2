```python
# Initial imports.
import pandas as pd
from path import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from imblearn.ensemble import EasyEnsembleClassifier
import seaborn as sns
```

```python
# Import dataset
wine_df = pd.read_csv(Path("resources/wine2.csv"),index_col=0)
wine_df.head()
```

```python
# Drop duplicate entries
wine_df= wine_df.drop_duplicates()
wine_df.head()
```

```python
wine_df.loc[wine_df["type"] == 0, "type"] = 'red'
wine_df.loc[wine_df["type"] == 1, "type"] = 'white'
wine_df.tail()
```

```python
sns.violinplot(data=wine_df, x="quality", y="sulphates", hue="type", split=True, palette=["lightcoral", "mediumturquoise"])
```

```python
sns.violinplot(data=wine_df, x="quality", y="alcohol", hue="type", split=True, palette=["lightcoral", "mediumturquoise"])
```

```python
sns.violinplot(data=wine_df, x="quality", y="total sulfur dioxide", hue="type", split=True, palette=["lightcoral", "mediumturquoise"])
```

```python
sns.violinplot(data=wine_df, x="quality", y="chlorides", hue="type", split=True, palette=["lightcoral", "mediumturquoise"])
```

```python
wine_df.loc[wine_df["type"] == 'red', "type"] = '0'
wine_df.loc[wine_df["type"] == 'white', "type"] = '1'
wine_df.tail()
```

```python
# Define the features set.
X = wine_df.copy()
X = X.drop("type", axis=1)
X.head()
```

```python
# Define the target set.
y = wine_df['type']
y[:5]
```

```python
# Splitting into Train and Test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)

# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
predictions
```

```python
# Calculating the confusion matrix.
cm = confusion_matrix(y_test, predictions)

# Create a DataFrame from the confusion matrix.
cm_df = pd.DataFrame(
    cm, index=["Actual Red", "Actual White"], columns=["Predicted Red", "Predicted White"])

cm_df
```

```python
# Calculating the accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Wine Type Prediction")
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```

```python
# Sort the features by their importance.
print("Wine Type Prediction Feature Importance")
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```
